```python
def my_fun2(x):
  return x**2 + 2*x+1
```

미분해보자. 아래는 미분 식

```python
def numerical_derivative(f, x):
  delta=1e-5
  return (f(x+delta)-f(x-delta))/(2*delta)
```

코드로 실제 미분을 해보자
y=x2 -> y`=2x 이니까 3들어가면 6

```python
result = numerical_derivative(my_fun1, 3.0)
print(result)
```

y=x^2 + 2x+1 
y` = 2x + 2  

```python
result2 = numerical_derivative(my_fun2, 2)
print(result2)
```

미분 값이 잘 나와. 미분 수식으로 적당하는 것을 확인 할 수 있어.

변수가 여러개야....................................
하나가 아니다. ㅇㅁㅇ......
변수가 하나가 아니니까.
1차원 반복자로 풀어서 사용.

```python
def numerical_derivative2(f,x):
  delta_x=1e-4
  grad=np.zeros_like(x) #0로 채워진 배열을 만들어

  it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite']) #읽,쓰반복자를 만든다.
  while not it.finished: #반복자가 끝이 아닌동안 반복한다.
    idx = it.multi_index
    tem_val = x[idx]
    x[idx]=tem_val+delta_x
    fx1=f(x)
    x[idx]=tem_val-delta_x
    fx2=f(x)
    grad[idx]=(fx1-fx2)/(2*delta_x)

    x[idx]=tem_val
    it.iternext()

  return grad
```

input 이라는 변수를 만들고

```python
def func3(input):
  x=input[0]
  y=input[1]

  return(2*x + 3*x*y + np.power(y,3)) #2x+3xy+y^3
```

미분을 해보자
y=2x + 3xy + y**3

y로 미분하면 3x + 3y**2
x로 미문하면 2 + 3y


```python
result3 = numerical_derivative2(func3, np.array([1.0, 2.0]))
print(result3)
```

```python
result4 = numerical_derivative2(my_fun1, np.array(3.0))
print(result4)
```
