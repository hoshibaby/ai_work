조금전까지는 위치를 찾은 것.
Top-k 가장 큰 k개 값과 index를 찾아줘.
각 행에서 제일 큰 값을 찾아줘.

```python
values, indes=torch.topk(x, k=1, dim=-1)
print(values)
print(indes)
print(values.size())
print(indes.size())
```

k=2 하면 큰거 2개를 찾아.
dim=-1 이걸 조절하면 차원 요소에서 찾을 수 있어.

```python
values, indes=torch.topk(x, k=2, dim=-1)
print(values)
print(indes)
print(values.size())
print(indes.size())
```

Masked Fill
