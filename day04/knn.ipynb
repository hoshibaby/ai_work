스케일 조정

```python
from sklearn.preprocessing import StandardScaler

s=StandardScaler() #값의 크기를 스케일 해주는데, 평균은 0, 표준편차는 1이 됨.count

x_train1=s.fit_transform(x_train)
x_test1=s.fit_transform(x_test)
print(x_train1)
#연산이 되었을때 작은 값이 될 수 있게 해줘
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10) #가장가까운 점 50개를 보고 분류함.
knn.fit(x_train1, y_train) #모델훈련을 시킴

#n_neighbors의 수치를 낮추면 정확도가 올라가
#결과는 아래처럼 나온다는데?
```

```python
from sklearn.metrics import accuracy_score # 정확도 측정은 사이키런안에 있어
y_pred=knn.predict(x_test1)

print(f'정확도 {accuracy_score(y_test, y_pred)}') #y_test는 정답 정확도가 0.76
```

함수가지고 와서 그대로 쓰는게 중요!!!!!!!!!!!!!


```python
k=50 #k를 1~10까지 줬을 때
acc_array = np.zeros(k)
for k in np.arange(1, k+1): #k가 작으면 작을 수록 정확하게 찾아. 정확도 제일 높은 값 찾기
  classifier=KNeighborsClassifier(n_neighbors=k).fit(x_train1, y_train)
  y_pred=classifier.predict(x_test1)
  acc=metrics.accuracy_score(y_test, y_pred)
  acc_array[k-1]=acc

max_acc=np.amax(acc_array)   #배열에서 가장 큰 값 찾아
acc_list = list(acc_array)
k=acc_list.index(max_acc)
print(f'정확도 {max_acc}으로 최적의 k는 {k+1}입니다')
```
