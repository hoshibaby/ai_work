```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# ===== [A] DATASET LOADER =====
from sklearn.datasets import load_wine

dataset = load_wine()   # üëà Ïó¨Í∏∞Îßå Î∞îÍæºÎã§ (load_iris, load_breast_cancer Îì±)
```

Ïù¥Î¶Ñ Î™ÖÎ™Ö, ÌîÑÎ†àÏûÑ ÎßåÎì§Í∏∞

```python
X = dataset.data
y = dataset.target
feature_names = dataset.feature_names
target_names = dataset.target_names if hasattr(dataset, 'target_names') else None

df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

display(df.head())

print("X shape:", X.shape)
print("y shape:", y.shape)
print("feature Í∞úÏàò:", len(feature_names))
print("class Í∞úÏàò:", len(set(y)))
print("target Î∂ÑÌè¨:\n", df['target'].value_counts().sort_index())

if target_names is not None:
    print("target names:", target_names)
```

```python
# ===== [C] TENSOR CONVERSION =====
x_data = torch.FloatTensor(X)
y_data = torch.LongTensor(y)

print("x_data:", x_data.shape, x_data.dtype)
print("y_data:", y_data.shape, y_data.dtype)
```

# ===== [D] MODEL & TRAIN =====

```python
input_dim = x_data.shape[1]
num_classes = len(torch.unique(y_data))

model = nn.Linear(input_dim, num_classes)
optimizer = optim.SGD(model.parameters(), lr=0.001)

epochs = 10000
for epoch in range(epochs):
    logits = model(x_data)
    loss = F.cross_entropy(logits, y_data)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 1000 == 0:
        pred = torch.argmax(logits, dim=1)
        acc = (pred == y_data).float().mean().item()
        print(f"epoch:{epoch:5d} loss:{loss.item():.4f} acc:{acc:.4f}")
```
